# Check Normal Distribution Using QQ Plot-Quantile â€“ Quantile plot

```python
import scipy.stats as stat
import pylab
import numpy as np
```

```python
import seaborn as sns
```

```python
df= sns.load_dataset('iris')
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
### if you want to check whether feature is gaussion or normal distribution
### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,16))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde = True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data(df,'sepal_width')
```

```python
plot_data(df,'petal_width')
```

```python
plot_data(df,'sepal_length')
```

```python
plot_data(df,'petal_length')
```
